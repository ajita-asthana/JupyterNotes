```python
#Performing data cleaning
T = temp.copy()
T.head()
```

```python
#Renaming the duplicate Month columns to M1 and M2 
d = {'Month': ['M1', 'M2']}
T = T.rename(columns=lambda c: d[c].pop(0) if c in d.keys() else c)
T.head()
```

```python
#Dropping one of the duplicate columns M2 to avoid discrepant calculations
T.drop(['M2'], axis=1, inplace=True)
T.head()
```

```python
#Renaming the column back to Month for better understanding
T.rename({'M1': 'Month'}, axis=1, inplace=True)
T
```

```python
temp = T.copy()
temp.head()
```

```python
# a. How many NaN values exist? Write the code that shows a single number that enumerates a count of the NaN values.
temp.isna().sum().sum()
```

```python
#How many NaN values are in each column? Write the code that enumerates a count of the NaN values.
temp_cols = temp.columns[temp.isnull().any()]
temp_cols
temp.isna().sum()
```

```python
#dropping the state, unnamed 8 and unnamed 9 columns as they contain only NaN values
temp.drop(['State', 'Unnamed: 8','Unnamed: 9'], axis=1, inplace=True)
temp.head()
```

```python
temp.iloc[178108:178111]
```

```python
#Row 178109 of the dataframe contains NaN values for City, Month, Day and Year columns 

#For City Column, we add the value as Lagos since the above and below rows have same value
temp['City'] = temp['City'].replace(np.nan, 'Lagos')
```

```python
#For the Day column we add value as 2 since the above and below row values are 1 and 3 respectively
temp.loc[178109, 'Day'] = 2
```

```python
#For the year column, both above and below row have values 2008, we will continue with this value
temp['Year'] = temp['Year'].replace(np.nan, 2008)
```

```python
#For NaN values in the Month column we notice the continuation of data from the previous rows. We will interpolate these values to fill the NaN values in the month column
temp['Month'].interpolate(method ='pad', limit_direction ='forward' ,limit =1,inplace=True)
```

```python
# To handle the value in the day column for row 1020004, we notice the missing value is 31 for the month 12, as the previous row is 30 for month 12
temp.loc[1020004, 'Day'] = 31
```
