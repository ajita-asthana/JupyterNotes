##Part 2: Building a class, method and objects

```
class crypto:

  def __init__(self, name, symbol, price, circulating_supply):
    print("You have created an object of type crypto!")
    self.name = name
    self.symbol = symbol
    self.price = price
    self.circulating_supply = circulating_supply
    print("The name is ", self.name)
    print("The symbol is ", self.symbol)
    print("The price is ", self.price)
    print("The circulating supply is ", self.circulating_supply)

  def setPrice(self, price):
    self.price = price
    print("The price is ",self.price)

  def setCirculating(self, circulating_supply):
    self.circulating_supply = circulating_supply
    print("The circulating supply is ", self.circulating_supply)

  def getMarketCap(self):
    marketCap =  self.price * self.circulating_supply
    print("The market cap for cryptocurrency", self.name,"is $",marketCap)
```

```
bitcoinObj = crypto("Bitcoin", "BTC", 19193.00, 19187300)
```

```
ethereumObj = crypto("Ethereum", "ETH", 1309.74, 122373863)
```

```
dogecoinObj = crypto("Dogecoin", "DOGE", 0.05933, 132670764300)
```

```
tetherObj = crypto("Tether", "USDT", 1.00, 68453108030)
```
